```python
import sys
import os
import importlib.util

# config.py の絶対パス
config_file_path = "/content/drive/MyDrive/kaggle/binary_class/src/config.py"
module_name = "config"

# 1. ファイルが存在するか最終確認
if not os.path.exists(config_file_path):
    print(f"エラー: 指定されたファイル '{config_file_path}' が見つかりません。パスを確認してください。")
else:
    # 2. ファイルパスからモジュール仕様を作成
    spec = importlib.util.spec_from_file_location(module_name, config_file_path)

    # 3. モジュール仕様からモジュールをロード
    config = importlib.util.module_from_spec(spec)

    # 4. モジュールを実行
    spec.loader.exec_module(config)

    # ロードされたモジュールがsys.modulesに追加されることを確認
    sys.modules[module_name] = config

    print(f"モジュール '{module_name}' が正常にロードされました。")

# これで'config'モジュールが使えるようになります
print(f"config.py の内容はこれです: {dir(config)}")
```
